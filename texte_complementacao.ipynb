```python
import os
import ssl
import httpx
from dotenv import load_dotenv
from groq import Groq

load_dotenv()

def criar_cliente_groq_ssl_desabilitado():
# Criar um contexto SSL que não verifica certificados
    ssl_context = ssl.create_default_context()
    ssl_context.check_hostname = False
    ssl_context.verify_mode = ssl.CERT_NONE

    http_client = httpx.Client(verify=False)

    client = Groq(
        api_key=os.getenv('GROQ_API'),
        http_client=http_client
    )

    return client

# Testar o cliente
client = criar_cliente_groq_ssl_desabilitado()
print("cliente criado com sucesso!")
```

### Complemento Textual

- Expansão de tópicos -> desenvolvimento de pontos em parágrafos
- Expansão criativa -> narrativas envolventes
- Expansão argumentativa -> justifictivas sólidas
- Expasão técnica -> explicações detalhadas e exemplos práticos
- Expansão de outiline -> transformação de estrutur em conteúdo completo

### Expansão de tópicos

```python
system_prompt = """
Você é um especialista em municação corporativa.
Expanda tópicos curtos em parágrafos claros que:
1) Desenvolvam cada ideia com exemplos reais
2) Mantenha flinidez e coetência
3) Sejam úteis para profissionais em cargos de liderança
"""

user_prompt = """
Expanda esses tópios sobre liderança moderna:

- Comunicação transparente
- Gestão deqeuipes híbridas
- Incentivo à inovação
"""

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.6
)
print(f"Resposta {response.choices[0].message.content}\n")
```
