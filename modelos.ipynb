### Modelo conecção do client

```python
import os
import ssl
import httpx
from dotenv import load_dotenv
from groq import Groq

load_dotenv()

def criar_cliente_groq_ssl_desabilitado():
# Criar um contexto SSL que não verifica certificados
    ssl_context = ssl.create_default_context()
    ssl_context.check_hostname = False
    ssl_context.verify_mode = ssl.CERT_NONE

    http_client = httpx.Client(verify=False)

    client = Groq(
        api_key=os.getenv('GROQ_API'),
        http_client=http_client
    )

    return client

# Testar o cliente
client = criar_cliente_groq_ssl_desabilitado()
print("cliente criado com sucesso!")
```

### Modelo de Prompt

```python
system_prompt = """

"""

user_prompt = """

"""

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.6
)
print(f"Resposta {response.choices[0].message.content}\n")
```
