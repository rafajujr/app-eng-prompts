```python
import os
import ssl
import httpx
from dotenv import load_dotenv
from groq import Groq

load_dotenv()

def criar_cliente_groq_ssl_desabilitado():
# Criar um contexto SSL que não verifica certificados
    ssl_context = ssl.create_default_context()
    ssl_context.check_hostname = False
    ssl_context.verify_mode = ssl.CERT_NONE

    http_client = httpx.Client(verify=False)

    client = Groq(
        api_key=os.getenv('GROQ_API'),
        http_client=http_client
    )

    return client

# Testar o cliente
client = criar_cliente_groq_ssl_desabilitado()
print("cliente criado com sucesso!")
```

### Sumarização Básica

- Tendência e dasafios no desenvolvimento de jogos indie

```python
system_prompt = """
Você é um editor técnico especializado em jogos independentes.
Crie um resumo objetivo como aproximadamente 1/3 do tamanho do texto original,
preservando os pontos centrais e o tom informativo.
"""

user_prompt = """
As Guerras Mundiais foram os eventos mais transformadores da era moderna, moldando a geopolítica, 
a tecnologia e a sociedade como as conhecemos hoje. A Primeira Guerra Mundial, que durou de 1914 a 1918, 
surgiu de uma mistura explosiva de nacionalismo extremo, disputas por colônias e alianças militares rígidas na Europa. 
O assassinato do herdeiro do trono austro-húngaro foi o pavio que acendeu o conflito, 
levando o mundo a uma guerra de desgaste sem precedentes. Diferente das batalhas anteriores, 
esta foi marcada pela vida terrível nas trincheiras e pelo uso de novas tecnologias mortais, 
como o gás venenoso, tanques e aviões. Quando o conflito terminou com a derrota da Alemanha e de seus aliados, 
o Tratado de Versalhes impôs condições tão severas aos derrotados que acabou criando um terreno 
fértil para o ressentimento e a instabilidade econômica.
"""

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.6
)
print(f"Resposta {response.choices[0].message.content}\n")
```

### Sumarização em Bullets

```python
system_prompt = """
você é um analista de gestão. gere um resume em Bullet points claros e hieráquicos:
- Destaque resultados (métricas)
- Liste problemas prioritários
- Sugira 3 ações imediatas e 2 ações estruturais
"""

user_prompt = """
Relatório piloto: Uma igreja adiquiriu um aplicação de gestão de mebros, mas não tem usado as funcionalidades do sistema
Resultados reliminates: foi formado uma equie para gestão e implentação das funcionalidades em cada ministerio da igreja
Observações: desafio mudar a mentalidade dos membros para uso; dar ideias para os lideres de cada ministerio sem passa a 
ideia que vai destitui-lo do cargo
"""

response = client.chat.completions.create(
    model="openai/gpt-oss-120b",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
    temperature=0.6
)
print(f"Resposta {response.choices[0].message.content}\n")
```
